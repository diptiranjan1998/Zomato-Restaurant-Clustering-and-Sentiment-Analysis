### Here we can see that the data type of Cost and Rating column is Object which is not valid. So we are gonna convert it to integer type.

```python
df.rating= df.rating.apply(lambda x: pd.to_numeric(x, errors= 'coerce'))
```

```python
df.cost= df.cost.apply(lambda x: pd.to_numeric((x.replace(',','')), errors= 'coerce'))
```

```python
df.collections.fillna('Not Available', inplace= True)
df.review.fillna('Not Available', inplace= True)
df.reviewer.fillna('Not Available', inplace= True)
df.metadata.fillna('0 Review , 0 Followers', inplace= True)
df.rating.fillna(df.loc[~df.rating.isna(),'rating'].median(), inplace= True)
```
